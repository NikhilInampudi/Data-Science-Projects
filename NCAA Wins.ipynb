```python
import pandas as pd

from matplotlib import pyplot as plt

df = pd.read_csv(r"C:\Users\nikhi\Downloads\cbb.csv")

conf = df.groupby(by='CONF')['W'].mean()

sorted_teams = conf.sort_values(ascending=False)

sorted_teams

top_teams = sorted_teams.head()

top_teams

figure = plt.figure(figsize=(10, 5))

plt.bar(x=df.CONF, height=df.W, color="Green")

plt.title("Top 5 Teams in NCAA History")
plt.xlabel("Teams")
plt.ylabel("Average Wins")
plt.xticks(rotation=90)

plt.show()
```

```python
#Finding top performing teams in NCAA history

conf_wins = df.groupby('CONF')['W'].sum()

conf_wins = conf_wins.reset_index()

conf_wins = conf_wins.sort_values(by='W', ascending=False)

filtered_conf = conf_wins[conf_wins['CONF'].isin(['ACC', 'SEC'])]

filtered_conf

plt.bar(x=conf_wins.CONF, height=conf_wins.W, color="Black")

plt.title("Total Conference Wins")
plt.xlabel("Conferences")
plt.ylabel("Total Wins")
plt.xticks(rotation = 90)

plt.show()
```
